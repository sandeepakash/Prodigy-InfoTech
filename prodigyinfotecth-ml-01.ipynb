```python
# importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Initializing and declaring the required datasets
train_features=pd.read_csv('/kaggle/input/house-price-prediction-challenge-machine-hack/Train.csv')
test_features=pd.read_csv('/kaggle/input/house-price-prediction-challenge-machine-hack/Train.csv')

train_features=train_features.iloc[:,:-1]
test_features=test_features.iloc[:,:-1]
train_target=train_features.iloc[:,-1]
test_target=test_features.iloc[:,-1]
```

```python
selected_columns = ['UNDER_CONSTRUCTION', 'RERA', 'BHK_NO.', 'SQUARE_FT', 'READY_TO_MOVE', 'RESALE', 'LONGITUDE', 'LATITUDE']
train_features = train_features.loc[:, selected_columns]
test_features=test_features.loc[:,selected_columns]
```

```python
plt.scatter(y=train_features['BHK_NO.'],x=train_target)
plt.show()
```

```python
train_features.shape,test_target.shape
```

```python
#using Standard sclar for improving the standardisation
from sklearn.preprocessing import StandardScaler
SC=StandardScaler()
SC.fit_transform(train_features,test_features)
```

```python
# Using Lasso Regression for Better prediction
from sklearn.linear_model import Lasso
model=Lasso().fit(train_features,train_target)
```

```python
#predicting from test data
target_pred=model.predict(test_features)
```

```python
# Evauating Metrics
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
MSE=mean_squared_error(test_target,target_pred)
MAE=mean_absolute_error(test_target,target_pred)
r2=r2_score(test_target,target_pred)
print(f"Mean squared error:{MSE} \n Mean Absolute Error:{MAE} \n R squared score:{r2}")
```

Thus the accuracy score is great and the model performs well in training and testing the data.Using the Lasso Regression model helps to improve the scores and predictive capabilities towards finding the model predictions,
